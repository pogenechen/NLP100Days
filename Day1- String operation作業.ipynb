## 讀入所需文件

```python
import pandas as pd
#TSV文件有別於CSV的文件是使用\t作为分隔符，CSV則是使用,作为分隔符。
dataset = pd.read_csv(r'Restaurant_Reviews.tsv', sep='\t')
```

```python
all_review = dataset['Review'].values
```

---

### 計算有多少個句子是以 . 結尾

```python
n = len(dataset.Review[dataset.Review.str.endswith('.')])

print('There are {} sentences end with .'.format(n))
```

### 將所有. 換成 ,

```python
dataset.Review = dataset.Review.str.replace('.',',')
all_review = dataset.values
```

```python
all_review
```

### 將所有sentence 中的 the 置換成 The

```python
dataset.Review = dataset.Review.str.replace('the','The')
all_review = dataset.values
```

```python
all_review
```

### 將偶數句子全部轉換為大寫，基數句子全部轉換為小寫

```python
dataset.Review = dataset.Review.map(lambda x:x.upper() if list(dataset.Review).index(x) % 2 == 1 else x.lower())
all_review = dataset.values
```

```python
all_review
```
